#### cleaning data

```python
import re

def cleanResume(txt):
    cleanText = re.sub(r'http\S+\s', ' ', txt)
    cleanText = re.sub(r'RT|cc', ' ', cleanText)
    cleanText = re.sub(r'#\S+\s', ' ', cleanText)
    cleanText = re.sub(r'@\S+', ' ', cleanText)
    cleanText = re.sub(r'[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^-`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText)  # Fixed this line
    cleanText = re.sub(r'\s+', ' ', cleanText)  # Fixed this line
    return cleanText
```

```python
cleanResume("my #### # $ #priyanshu website like is this http://heloword and access it @gmail.com")
```

```python
df['Resume']=df['Resume'].apply(lambda x: cleanResume(x))
```

```python
df['Resume'][0]
```

#### word into categorical values

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
```

```python
df.Category.unique()
```

#### vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')  # Corrected name
tfidf.fit(df['Resume'])
requiredtext = tfidf.transform(df['Resume'])
```

```python
df
```

```python
requiredtext
```

#### Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(requiredtext,df['Category'],test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

#### now lets train the model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

clf =  OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train,y_train)
ypred = clf.predict(X_test)
print(accuracy_score(y_test,ypred))
```

#### prediction system

```python
import pickle
pickle.dump(tfidf,open('tfidf.pkl','wb'))
pickle.dump(clf,open('clf.pkl','wb'))
```
